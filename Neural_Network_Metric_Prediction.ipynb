```python
from re import A
import pandas as pd
import numpy as np
import random
import math
import matplotlib.pyplot as plt


df = pd.read_csv('dataset.csv')
k = len(df)
altitude = np.zeros(k)
indicated_airspeed = np.zeros(k)
pitch = np.array(k)
roll = np.array(k)
temp1 = df["altitude"].to_numpy()
temp2 = df["indicated_airspeed"].to_numpy()
temp3 = df["pitch"].to_numpy()
temp4 = df["roll"].to_numpy()
altitude = temp1.astype(float)
original_altitude = altitude
#altitude = np.append(0,altitude)
indicated_airspeed = temp2.astype(float)
original_speed = indicated_airspeed
#indicated_airspeed = np.append(0,indicated_airspeed)
pitch = temp3.astype(float)
original_pitch = pitch
next_pitch = temp3.astype(float)
next_pitch = np.append(pitch,0)
roll = temp4.astype(float)
original_roll = roll
#roll = np.append(0,roll)
weights = np.random.uniform(0.9,1,40)
weights = np.absolute(weights)
partial_weights = np.empty_like(weights)
L = 0.000001
alpha = 0.01
N = 93000
mse_list = np.zeros(N)
n_list = np.zeros(N)



min = float(altitude.min())
max = float(altitude.max())
altitude = (altitude - min)/(max - min)
altitude_min = min
altitude_max = max

min = float(indicated_airspeed.min())
max = float(indicated_airspeed.max())
indicated_airspeed = (indicated_airspeed - min)/(max - min)
indicated_airspeed_min = min
indicated_airspeed_max = max

min = float(pitch.min())
max = float(pitch.max())
pitch = (pitch - min)/(max - min)
next_pitch = (next_pitch - min)/(max - min)
pitch_min = min
pitch_max = max


min = float(roll.min())
max = float(roll.max())
roll = (roll - min)/(max - min)
roll_min = min
roll_max = max
print(altitude)
print(indicated_airspeed)
print(roll)
print(pitch)

for n in range(N):
  print("Iteration: ",n)

  in_h1 = weights[0] * indicated_airspeed + weights[8] * altitude + weights[16] * roll + weights[24] * pitch
  in_h2 = weights[1] * indicated_airspeed + weights[9] * altitude + weights[17] * roll + weights[25] * pitch
  in_h3 = weights[2] * indicated_airspeed + weights[10] * altitude + weights[18] * roll + weights[26] * pitch
  in_h4 = weights[3] * indicated_airspeed + weights[11] * altitude + weights[19] * roll + weights[27] * pitch
  in_h5 = weights[4] * indicated_airspeed + weights[12] * altitude + weights[20] * roll + weights[28] * pitch
  in_h6 = weights[5] * indicated_airspeed + weights[13] * altitude + weights[21] * roll + weights[29] * pitch
  in_h7 = weights[6] * indicated_airspeed + weights[14] * altitude + weights[22] * roll + weights[30] * pitch
  in_h8 = weights[7] * indicated_airspeed + weights[15] * altitude + weights[23] * roll + weights[31] * pitch

  out_h1 = np.maximum(in_h1,0)
  out_h2 = np.maximum(in_h2,0)
  out_h3 = np.maximum(in_h3,0)
  out_h4 = np.maximum(in_h4,0)
  out_h5 = np.maximum(in_h5,0)
  out_h6 = np.maximum(in_h6,0)
  out_h7 = np.maximum(in_h7,0)
  out_h8 = np.maximum(in_h8,0)


  in_o1 = weights[32] * out_h1 + weights[33] * out_h2 + weights[34] * out_h3 + weights[35] * out_h4 + weights[36] * out_h5 + weights[37] * out_h6 + weights[38] * out_h7 + weights[39] * out_h8

  out_o1 = in_o1


  sig_o1 = 1
  sig_h1 = np.where(in_h1 < 0, 0, 1)
  sig_h2 = np.where(in_h2 < 0, 0, 1)
  sig_h3 = np.where(in_h3 < 0, 0, 1)
  sig_h4 = np.where(in_h4 < 0, 0, 1)
  sig_h5 = np.where(in_h5 < 0, 0, 1)
  sig_h6 = np.where(in_h6 < 0, 0, 1)
  sig_h7 = np.where(in_h7 < 0, 0, 1)
  sig_h8 = np.where(in_h8 < 0, 0, 1)

  error_o1 = out_o1 - next_pitch[1:]
  delta_o1 = error_o1 * sig_o1

  error_h1 = delta_o1 * weights[32]
  delta_h1 = error_h1 * sig_h1

  error_h2 = delta_o1 * weights[33]
  delta_h2 = error_h2 * sig_h2

  error_h3 = delta_o1 * weights[34]
  delta_h3 = error_h3 * sig_h3

  error_h4 = delta_o1 * weights[35]
  delta_h4 = error_h4 * sig_h4

  error_h5 = delta_o1 * weights[36]
  delta_h5 = error_h5 * sig_h5

  error_h6 = delta_o1 * weights[37]
  delta_h6 = error_h6 * sig_h6

  error_h7 = delta_o1 * weights[38]
  delta_h7 = error_h7 * sig_h7

  error_h8 = delta_o1 * weights[39]
  delta_h8 = error_h8 * sig_h8

  partial_weights[0] = indicated_airspeed.T.dot(delta_h1)
  partial_weights[1] = indicated_airspeed.T.dot(delta_h2)
  partial_weights[2] = indicated_airspeed.T.dot(delta_h3)
  partial_weights[3] = indicated_airspeed.T.dot(delta_h4)
  partial_weights[4] = indicated_airspeed.T.dot(delta_h5)
  partial_weights[5] = indicated_airspeed.T.dot(delta_h6)
  partial_weights[6] = indicated_airspeed.T.dot(delta_h7)
  partial_weights[7] = indicated_airspeed.T.dot(delta_h8)
  partial_weights[8] = altitude.T.dot(delta_h1)
  partial_weights[9] = altitude.T.dot(delta_h2)
  partial_weights[10] = altitude.T.dot(delta_h3)
  partial_weights[11] = altitude.T.dot(delta_h4)
  partial_weights[12] = altitude.T.dot(delta_h5)
  partial_weights[13] = altitude.T.dot(delta_h6)
  partial_weights[14] = altitude.T.dot(delta_h7)
  partial_weights[15] = altitude.T.dot(delta_h8)
  partial_weights[16] = roll.T.dot(delta_h1)
  partial_weights[17] = roll.T.dot(delta_h2)
  partial_weights[18] = roll.T.dot(delta_h3)
  partial_weights[19] = roll.T.dot(delta_h4)
  partial_weights[20] = roll.T.dot(delta_h5)
  partial_weights[21] = roll.T.dot(delta_h6)
  partial_weights[22] = roll.T.dot(delta_h7)
  partial_weights[23] = roll.T.dot(delta_h8)
  partial_weights[24] = pitch.T.dot(delta_h1)
  partial_weights[25] = pitch.T.dot(delta_h2)
  partial_weights[26] = pitch.T.dot(delta_h3)
  partial_weights[27] = pitch.T.dot(delta_h4)
  partial_weights[28] = pitch.T.dot(delta_h5)
  partial_weights[29] = pitch.T.dot(delta_h6)
  partial_weights[30] = pitch.T.dot(delta_h7)
  partial_weights[31] = pitch.T.dot(delta_h8)
  partial_weights[32] = out_h1.T.dot(delta_o1)
  partial_weights[33] = out_h2.T.dot(delta_o1)
  partial_weights[34] = out_h3.T.dot(delta_o1)
  partial_weights[35] = out_h4.T.dot(delta_o1)
  partial_weights[36] = out_h5.T.dot(delta_o1)
  partial_weights[37] = out_h6.T.dot(delta_o1)
  partial_weights[38] = out_h7.T.dot(delta_o1)
  partial_weights[39] = out_h8.T.dot(delta_o1)

  weights = weights - (L * partial_weights)

  y = np.power((out_o1 - next_pitch[1:]), 2)
  mse = (y.sum())/k

  rmse = math.sqrt(mse)
  #print(in_h1)
  print(out_o1)
  print(pitch)
  print("MSE: ",mse)


  mse_list[n] = mse
  n_list[n] = n

plt.plot(n_list, mse_list,marker='o',color="blue", linewidth=1)

plt.xlabel('Iterations')
plt.ylabel('MSE')
plt.title("Loss and plot at each iteration N")
plt.legend(["Iteration"], loc="upper right")
plt.savefig('Loss_plot.png')
plt.xlabel('Iterations')
plt.ylabel('MSE')
plt.title("Loss and plot at each iteration N")
plt.legend(["Iteration"], loc="upper right")
plt.savefig('Loss_plot.png')

x1 = 1


while(x1 ==1):
  i = int(input("Enter the index to use to predict index+1"))

  speed1 = indicated_airspeed[i]
  altitude1 = altitude[i]
  roll1 = roll[i]
  pitch1 = pitch[i]
  print("The speed is: ", original_speed[i])
  print("The altitude is: ", original_altitude[i])
  print("The roll is: ", original_altitude[i])
  print("The pitch is: ", original_pitch[i])

  h1_in = weights[0] * speed1 + weights[8] * altitude1 + weights[16] * roll1 + weights[24] * pitch1
  h2_in = weights[1] * speed1 + weights[9] * altitude1 + weights[17] * roll1 + weights[25] * pitch1
  h3_in = weights[2] * speed1 + weights[10] * altitude1 + weights[18] * roll1 + weights[26] * pitch1
  h4_in = weights[3] * speed1 + weights[11] * altitude1 + weights[19] * roll1 + weights[27] * pitch1
  h5_in = weights[4] * speed1 + weights[12] * altitude1 + weights[20] * roll1 + weights[28] * pitch1
  h6_in = weights[5] * speed1 + weights[13] * altitude1 + weights[21] * roll1 + weights[29] * pitch1
  h7_in = weights[6] * speed1 + weights[14] * altitude1 + weights[22] * roll1 + weights[30] * pitch1
  h8_in = weights[7] * speed1 + weights[15] * altitude1 + weights[23] * roll1 + weights[31] * pitch1

  if(h1_in <= 0):
    h1_out = 0
  else:
    h1_out = h1_in

  if(h2_in <= 0):
    h2_out = 0
  else:
    h2_out = h2_in

  if(h3_in <= 0):
    h3_out = 0
  else:
    h3_out = h3_in

  if(h4_in <= 0):
    h4_out = 0
  else:
    h4_out = h4_in

  if(h5_in <= 0):
    h5_out = 0
  else:
    h5_out = h5_in

  if(h6_in <= 0):
    h6_out = 0
  else:
    h6_out = h6_in

  if(h7_in <= 0):
    h7_out = 0
  else:
    h7_out = h7_in

  if(h8_in <= 0):
    h8_out = 0
  else:
    h8_out = h8_in

  o1_in = weights[32] * h1_out + weights[33] * h2_out + weights[34] * h3_out + weights[35] * h4_out + weights[36] * h5_out + weights[37] * h6_out + weights[38] * h7_out + weights[39] * h8_out
  o1_out = o1_in

  predicted = (o1_out*(pitch_max - pitch_min) )+ pitch_min


  print("The predicted value is: ", predicted)
  x1 = int(input("Enter 1 to predict or 0 to quit"))

```
